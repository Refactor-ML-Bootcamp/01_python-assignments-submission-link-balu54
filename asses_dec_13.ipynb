```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/scikit-learn-videos/master/data/Advertising.csv")
df.drop("Unnamed: 0",axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(),annot=True,cmap="Blues")
```

```python
sns.histplot(df)
```

```python
df.iloc[:,:-1].plot(kind='box', title='sales')
```

```python
# checking sales by tv, newspaper, radio
sns.pairplot(y_vars="Sales",x_vars=["Radio","Newspaper","TV"],data=df)
```

```python
sns.regplot(df.Sales,df.TV,scatter_kws={"color": "black"}, line_kws={"color": "red"})#,x_estimator=np.mean)
```

```python
sns.lmplot(x="Sales",y="TV",data=df,scatter_kws={"color": "black"}, line_kws={"color": "red"})
```

```python
sns.lineplot(data=df, x="Sales",y="Newspaper")
sns.lineplot(data=df, x="Sales",y="Radio")
sns.lineplot(data=df, x="Sales",y="TV")
```
